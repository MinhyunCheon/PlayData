## 과제 1 -> txt 파일안에 금액을 남성, 여성으로 구분해서 토탈값 구하기 
    -> 남성 : xxxxxxx원 
    -> 여성 : xxxxxxx원 
## 과제 2 -> txt파일 10000개를 남성, 여성폴더 생성하고 해당폴더에 파일 옮기기 
   -> 남성은 남성폴더에..
   -> 여성은 여성폴더에 파일 옮기기

```python
import os
import shutil

list_path = './이름_생년_성별_10000'
file_list = os.listdir(list_path)
total_man_money = 0
total_woman_money = 0

# 2번 문제 파일 분류
if not os.path.isdir("./man"): os.mkdir("./man")
if not os.path.isdir("./woman"): os.mkdir("./woman")

for file_name in file_list:
    now_file_path = f'{list_path}/{file_name}'
    with open(now_file_path, 'rb') as f:
        context = f.read()
        is_man = True
        
        # 1번 문제 성별 금액 출력
        if int(file_name.split('_')[2][0]) % 2 != 0:
            total_man_money += int(context.decode('utf-8').replace(',', ''))
            is_man = True
        else:
            total_woman_money += int(context.decode('utf-8').replace(',', ''))
            is_man = False

    # 2번 문제 파일 분류
    shutil.move(now_file_path, f'./{"man" if is_man else "woman"}/{file_name}')

# 1번 문제 성별 금액 출력
print(f'남성 : {format(total_man_money, ",d")}원')
print(f'여성 : {format(total_woman_money, ",d")}원')
            
            

```

```python
root = ET.fromstring(r.text)
```

```python
items = root.iter(tag = 'item')
```
