```python
# 1 ~ 5월 데이터 병함
sub_dataFrame = pd.DataFrame()
csv_list = os.listdir('./Subway')
for x in range(0, 5):
    temp = pd.read_csv(f'./Subway/{csv_list[x]}', encoding='euc-kr')
    temp.rename(columns={'구 분' : '구분'}, inplace=True)
    temp = temp.groupby(['역명', '날짜', '구분']).sum()
    if len(sub_dataFrame) == 0: sub_dataFrame = temp
    else: 
        sub_dataFrame = pd.concat([sub_dataFrame, temp])

# 6 ~ 12월
for x in range(5, 12):
    temp = pd.read_csv(f'./Subway/{csv_list[x]}', encoding='euc-kr')
    temp = temp.groupby(['역명', '날짜', '구분']).sum()
    temp.columns = [x.replace('~', ' ~ ') for x in temp.columns]
    temp.columns = [x.replace('시', '') for x in temp.columns]
    sub_dataFrame = pd.concat([sub_dataFrame, temp])
```
