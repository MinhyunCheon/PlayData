# Generator
* yield를 만나면 멈추고(메모리에 상주한 상태) \_\_next__()을 호출하면 진행

```python
def generator_func():
    n = 0
    while n <= 2:
        yield n
        n += 1
```

```python
generator_1 = generator_func()
```

```python
generator_1.__next__()
```

### for문은 __iter__ 메소드에서 자동으로 next()를 호출하는 형식으로 실행됨

```python
# 피보나치 수열 generator
def fib_generator():
    a, b = 0, 1
    while True:
        yield b
        a, b = b, a + b
```

```python
fib = fib_generator()
```

```python
fib.__next__()
```
