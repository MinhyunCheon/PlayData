# 강남역 데이터만 추출

```python
subway
```

```python
subway[subway['역명'].str.contains('강남')]
```

```python
# 2가지 이상 단어 포함 여부
subway[subway['역명'].str.contains('강남|서초')]
```

```python
gangnam = subway[subway['역명'].str.contains('강남')].copy()
```

```python
# 월 컬럼 생성
gangnam.날짜.apply(lambda x : x.split('-')[1])
```

```python
gangnam['월'] = gangnam.날짜.apply(lambda x : x.split('-')[1])
```

```python
# 월별 강남역 이용 현황 집계(승, 하차 통합)
# astype()로 컬럼의 자료형 변경 가능(메모리 최적화를 위해 사용)
gangnam.groupby('월').sum()
```

```python
gangnam_monthly_data = gangnam.groupby('월').sum()
```

```python
# 월별 합계
gangnam_monthly_data.apply(lambda x : sum(x), axis=1)
```

```python
gangnam_monthly_data['합계'] = gangnam_monthly_data.apply(lambda x : sum(x), axis=1)
```

```python
# 시간별 합계
gangnam_monthly_data.apply(lambda x : sum(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [16, 9]
```

```python
with plt.style.context('ggplot'):
    plt.bar(gangnam_monthly_data['합계'].index, gangnam_monthly_data['합계'].values, width=0.35, label='강남역')
    plt.title('강남역')
    plt.ylabel('승객수')
    plt.xlabel('월')
    plt.legend() # 우상단 라벨
    plt.rc('font', family='Malgun Gothic') # 판다스에서 기본적으로 한글을 사용하려면 오류가 발생하므로 별도의 폰트 설정 필요
    plt.show()
```

```python
# 월별 승차, 하차
gangnam.groupby(by=['월','구분']).sum().apply(lambda x : sum(x), axis=1)
```

```python
gangnam_io = pd.DataFrame(gangnam.groupby(by=['월','구분']).sum().apply(lambda x : sum(x), axis=1))
```

```python
gangnam_io = pd.DataFrame(gangnam_io, columns=['승객수']).reset_index()
```

```python
gangnam_io
```

```python
with plt.style.context('ggplot'):
    fig, ax = plt.subplots(1)
    
    rect1 = ax.bar(gangnam_io[gangnam_io.구분 == '승차']['월'], 
                   gangnam_io[gangnam_io.구분 == '승차']['승객수'], 
                   0.35, alpha = 0.4, color='b', label='승차')
    
    rect2 = ax.bar(gangnam_io[gangnam_io.구분 == '하차']['월'],
                   gangnam_io[gangnam_io.구분 == '하차']['승객수'],
                   0.35, 
                   bottom=gangnam_io[gangnam_io.구분 == '승차']['승객수'],
                   alpha = 0.4, color='r', label='하차')
    
    ax.set_xlabel("월")
    ax.set_ylabel("승객수")
    ax.set_title("월별 강남역 승하차")
```
