# 재귀함수

```python
def gcd(m, n):
    if m == 0:
        print(m)
        return
    if m % n == 0:
        print(n)
        return
    else:
        temp = m
        m = m % n
        n = temp
        print(f"{m}, {n}")
        gcd(m, n)
```

```python
gcd(20, 3)
```

```python
# 기념일 계산
a = [x for x in range(0, 7)]
b = ["월", "화", "수", "목", "금", "토", "일"]
day_dict = dict(zip(a, b))

def special_day(day):
    for x in range(100, 1001, 100):
        s_day = day + timedelta(days = x)
        print(f"{day.strftime('%Y-%m-%d')}로 부터 {x}일 뒤는 {s_day.strftime('%Y-%m-%d')} {day_dict[s_day.weekday()]}요일")
```

```python
special_day(datetime.now())
```
